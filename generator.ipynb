## Import Libraries

```python
import os 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

## Uniform Distribution

```python
a = 0
b = 10
N_values = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
xmin = 0
xmax = 10
for N in N_values:
    x = np.random.uniform(a, b, N)
    plt.title("N={}".format(N))
    sns.distplot(x, kde_kws={"color": "r", "lw": 3, "label": "KDE"})
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(xmin, xmax)
    plt.legend()
    plt.show()
```

## Normal Distribution

```python
mean = 0
scale = 0.5
```

```python
N_values = [10, 100, 1000, 10000, 100000, 1000000, 10000000]
xmin = 0
xmax = 10
for N in N_values:
    x = np.random.normal(mean, scale, N)
    plt.title("N={}".format(N))
    sns.distplot(x, kde_kws={"color": "r", "lw": 3, "label": "KDE"})
    plt.legend()
    plt.show()
```
